```python
# Import the libraries.
import os
import tweepy as tw
import pandas as pd
import numpy as np
import configparser

config = configparser.ConfigParser()
config.read('Config.ini')


# Assign your Twitter Developer API‚Äôs and Access Tokens to variables.
api_key = config['twitter']['api_key']
api_secret = config['twitter']['api_key_secret']

access_token = config['twitter']['access_token']
access_token_secret = config['twitter']['access_token_secret'] 


# Authenticate the API‚Äôs and Access Tokens.
auth = tw.OAuthHandler(api_key, api_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)

# Define the search words and start date as variables. 
search_words = "football"
date_since = "2020-01-01"


```
